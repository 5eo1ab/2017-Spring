테이블의 구조를 가늠해볼 수 있다.
미국의 주(state)는 52개로 구성되어 있기때문에 `state_abbreviation`은 52개의 유니크한 값을 갖고, state에 속하는 `area_name` record를 갖는다.
이때 `state_abbreviation`가 null인 변수는 state에 대한 총합? 일 것으로 추정.
제일 첫번째 레코드는 미국 전체의 결과일 것으로 추정된다.

가볍게 state level에서 데이터를 탐색해보고 싶기때문에 세부적인 위 결과에서 미국 전체의 결과를 제외한 데이터셋을 사용하도록 하자.

```python
fips_state = data[data[cols[2]].isnull()==True][cols[0]][1:]
abbr_state = {}
for f in fips_state:
    k = data[data[cols[0]]==f][cols[1]].values[0]
    if len(data[data[cols[0]]==f+1][cols[2]]) > 0:
        v = data[data[cols[0]]==f+1][cols[2]].values[0]
    else :
        continue
    abbr_state[k] = v
#abbr_state
```
